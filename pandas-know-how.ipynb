# pandas know how

- where
- left/right/inner join

```python
import pandas as pd
import numpy as np
from datetime import datetime
import codecs
```

```python
# 表示設定
pd.options.display.max_rows = 10
```

```python
RATING = pd.read_csv('ml-100k/u.data', delimiter='\t', header=None)
RATING.columns = ['user_id', 'movie_id', 'rating', 'timestamp']
```

```python
with codecs.open("ml-100k/u.item", "r", "UTF-8", "ignore") as file:
    MOVIE = pd.read_table(file, delimiter="|", header=None)
    MOVIE.columns = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
                    'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
                    "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama',
                    'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
                    'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
USER = pd.read_csv('ml-100k/u.user', delimiter='|', header=None)
USER.columns = ['user_id', 'age', 'gender', 'occupation', 'zipcode']
```

## where

```python
# 普通の検索
## select * from RATING where user_id = 196;
RATING[RATING.user_id == 196]
```

```python
# 複数検索
## select * from RATING where user_id = 100 and movie_id = 242;
RATING[(RATING.user_id == 196) & (RATING.movie_id == 242)]
```

```python
# sql っぽく検索
RATING.query('user_id == 196 & movie_id == 242')
```

```python
# もとの DataFrame のサイズを保ったまま検索
RATING.where(RATING.user_id == 196)
```

## left/right/inner join

```python
# left join
## select * from RATING left join MOVIE on RATING.movie_id = MOVIE.movie_id limit 5
pd.merge(RATING, MOVIE, how='left', left_on=RATING.movie_id, right_on=MOVIE.movie_id)
```

```python
# right join
## select * from RATING right join MOVIE on RATING.movie_id = MOVIE.movie_id limit 5
pd.merge(RATING, MOVIE, how='right', left_on=RATING.movie_id, right_on=MOVIE.movie_id)
```

```python
# inner join
## select * from RATING inner join MOVIE on RATING.item_id = MOVIE.movie_id limit 5
pd.merge(RATING, MOVIE, how='inner', left_on=RATING.item_id, right_on=MOVIE.movie_id)
```

## cross join

```python
# cross join
## user_id と movie_id の全ての組み合わせを作る
user_id = USER[['user_id']]
movie_id = MOVIE[['movie_id']]
user_id['key'] = 0
movie_id['key'] = 0

pd.merge(user_id, movie_id, on='key').drop('key', axis=1)
```

## group by

```python
# ユーザーごとの rating の平均値
## select avg(rating) from RATING group by user_id;
RATING.groupby('user_id').mean()[['rating']]
```

## sort

```python
# user_id で昇順にソート
RATING.sort_values('user_id')
```

```python
# user_id で降順にソート
RATING.sort_values('user_id', ascending=False)
```

```python
# 複数のキーで並び替える
## timestamp で並び替えてから user_id で並び替える
RATING.sort_values(['timestamp', 'user_id'])
```
